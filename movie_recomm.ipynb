```python
import json

cnt = 900

while len(movie_dict) < n_sample:
    imdb_id = movie_links.imdbId[cnt]
    ap = 7-len(str(imdb_id))
    ap = ''.join([str(0) for i in range(ap)])
    imdb_id = f'tt{ap}{str(imdb_id)}'
    try:
        res = requests.get(f'{api_url}apikey={api_key}&i={imdb_id}', headers={'User-Agent': 'Chrome/61'})
        data = json.loads(res.text)
        if data['Response'] == 'True':
            movie_dict.update({data['imdbID']: data})
            cnt += 1
            print(f'{cnt} added.')
        else: 
            print(f'{cnt} err: '+data['Error'])
            cnt += 1
        sleep(.1)
    except Exception as e:
        print("Error: "+str(e))
        sleep(.1)

```

```python
# movies
import pandas as pd

movie_db = pd.DataFrame(columns={"imdb_id", "title", "year", "runtime", "countries", "languages", "released", "plot", "imdbrating", "imdbvote"})

mon_dict = {
    "Jan": 1,
    "Feb": 2,
    "Mar": 3,
    "Apr": 4,
    "May": 5,
    "Jun": 6,
    "Jul": 7,
    "Aug": 8,
    "Sep": 9,
    "Oct": 10,
    "Nov": 11,
    "Dec":12
}

for key, value in m_dict.items():
    [dd, mm, yr] = value["Released"].split(" ")

    movie_db = movie_db.append({
        "imdb_id": key,
        "title": value["Title"],
        "year": value["Year"],
        "runtime": value["Runtime"].split(' ')[0],
        "countries": value["Country"].split(', '), 
        "languages": value["Language"].split(', '),
        "released": f'{int(yr)}-{mon_dict[mm]:02}-{int(dd):02}',
        "plot": value["Plot"],
        "imdbrating": value["imdbRating"],
        "imdbvote": value["imdbVotes"].replace(",", '')
    }, ignore_index=True)

movie_db.to_csv("movie-db.csv", index=False)

movie_db
```

```python
# Actors
actor_dict = np.array([])

for key, value in movie_dict.items():
    actor_dict = np.append(actor_dict, value["Actors"].split(', '))

actor_dict = np.unique(actor_dict)
actor_db = pd.DataFrame(actor_dict, columns=["name"])
actor_db.index += 1
actor_db.to_csv("actor-db.csv", index_label="actor_id")
actor_db
```

```python
# Director
director_dict = np.array([])

for key, value in movie_dict.items():
    director_dict = np.append(director_dict, value["Director"].split(', '))

director_dict = np.unique(director_dict)
director_db = pd.DataFrame(director_dict, columns=["name"])
director_db.index += 1
director_db.to_csv("director-db.csv", index_label="director_id")
director_db
```

```python
# Users 
ratings = pd.read_csv('ml-latest-small/ratings.csv')

user_db = pd.DataFrame(np.unique(ratings.userId.to_numpy()), columns=["user_id"])
user_db.index += 1
user_db.to_csv('user-db.csv', index=False)
user_db
```

```python
# Rated relation
rated_db = pd.DataFrame(columns=["user_id", "imdb_id", "rating", "timestamp"])

#ratings must be initiated in this block
for row in ratings.index:
    movie_id = ratings.movieId[row]
    imdb_id = movie_links.loc[movie_links.movieId == movie_id]['imdbId'].to_numpy()[0]
    ap = 7-len(str(imdb_id))
    ap = ''.join([str(0) for i in range(ap)])
    imdb_id = f'tt{ap}{str(imdb_id)}'
    if imdb_id in m_dict:
        rated_db = rated_db.append({
            "user_id": ratings.userId[row],
            "imdb_id": imdb_id,
            "rating": ratings.rating[row],
            "timestamp": ratings.timestamp[row]
        }, ignore_index=True)

rated_db.index += 1
rated_db.to_csv('rated-db.csv', index=False)
rated_db
    
```

```python
# directed relationship
directed_db = pd.DataFrame(columns=["imdb_id", "director"])

for key, value in m_dict.items():
    for i in value['Director'].split(', '):
        directed_db = directed_db.append({
            'imdb_id': key,
            'director': i
        }, ignore_index=True)

directed_db.index +=1
directed_db.to_csv('directed-db.csv', index_label='idx')
directed_db
```

```python
# in_genre relationship
in_genre_db = pd.DataFrame(columns=["imdb_id", "genre"])

for key, value in m_dict.items():
    for i in value['Genre'].split(', '):
        in_genre_db = in_genre_db.append({
            'imdb_id': key,
            'genre': i
        }, ignore_index=True)

in_genre_db.index +=1
in_genre_db.to_csv('in_genre-db.csv', index_label='idx')
in_genre_db
```

```python
# acted_in relationship
acted_in_db = pd.DataFrame(columns=["imdb_id", "actor"])

for key, value in m_dict.items():
    for i in value['Actors'].split(', '):
        acted_in_db = acted_in_db.append({
            'imdb_id': key,
            'actor': i
        }, ignore_index=True)

acted_in_db.index +=1
acted_in_db.to_csv('acted_in-db.csv', index_label='idx')
acted_in_db
```
